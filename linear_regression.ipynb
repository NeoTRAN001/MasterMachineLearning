```python
import matplotlib.pyplot as plt
import matplotlib.lines as lines
import numpy as np

plt.style.use("seaborn-dark") # Por el background oscuro

def calcular_ganancia(x):
    y = 0.8 * x + 5
    return y

print(calcular_ganancia(0))
```

```python
x = np.array([0, 6.5, 13.2, 18.1, 24.9, 28.2, 36.1])
y = np.array([5, 10, 17, 20, 25, 30, 35])
```

```python
def showGrid():
    plt.figure(figsize=(7,7))
    plt.scatter(x, y, s=60)
    plt.ylim(bottom=0)
    plt.title("Valores")
    plt.xlabel("Dinero Invertido")
    plt.ylabel("Dinero Ganado")
    plt.grid()
    plt.show()
```

```python
showGrid()
```

```python
x2 = np.arange(0, 40)
y2 = 0.8 * x2 + 5
```

```python
plt.figure(figsize=(7,7))
plt.plot(x2, y2, color="red", label="Nuestra solution")
plt.scatter(x, y, s=60)
plt.ylim(bottom=0)
plt.title("Valores")
plt.xlabel("Dinero Invertido")
plt.ylabel("Dinero Ganado")
plt.grid()
plt.show()
```
